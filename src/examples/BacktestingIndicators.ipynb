```kotlin
import org.dgawlik.signals.EventFrequency
import org.dgawlik.signals.Timeline
import indicator.signals.org.dgawlik.signals.IndicatorConfig
import library.indicator.signals.org.dgawlik.signals.RSI
import ticker.signals.org.dgawlik.signals.Ticker
import java.time.LocalDateTime

var ticker = Ticker.defaultOf("AAPL");

var timeline = Timeline.forBacktesting(EventFrequency.ONE_DAY)
timeline.addSignal(ticker)

var indicatorConfig = IndicatorConfig.builder()
    .period(14)
    .build();
var rsi = RSI("AAPL", indicatorConfig)
timeline.addSignal(rsi)

var oneYearAgo = LocalDateTime.now().minusYears(1);
var now = LocalDateTime.now();

var eventsAg = timeline.getEvents(oneYearAgo, now, "AAPL", "RSI.AAPL")

val xTime = (0 until eventsAg.size.toLong()).toList()
val y1 = eventsAg.stream().map { it.forId("AAPL").getValue("close") }.toList()
val y2 = eventsAg.stream().map { it.forId("RSI.AAPL").getValue("primary") }.toList()

plot {
    line {
        x(xTime)
        y(y1)

        color = Color.BLUE
        width = 2.0
    }
    line {
        x(xTime)
        y(y2)

        color = Color.RED
        width = 2.0
    }
}
```
