```kotlin
import org.dgawlik.signals.EventFrequency
import org.dgawlik.signals.Timeline
import ticker.signals.org.dgawlik.signals.Ticker
import java.time.LocalDateTime

var tickers = Ticker.defaultTickers("AAPL", "TSLA")

var timeline = Timeline.forBacktesting(EventFrequency.ONE_DAY)
timeline.addSignals(tickers)

var oneYearAgo = LocalDateTime.now().minusYears(1);
var now = LocalDateTime.now();

var eventsAg = timeline.getEvents(oneYearAgo, now, "AAPL", "TSLA")

val xTime = (0 until eventsAg.size.toLong()).toList()
val y1 = eventsAg.stream().map { it.forId("AAPL").getValue("close") }.toList()
val y2 = eventsAg.stream().map { it.forId("TSLA").getValue("close") }.toList()

plot {
    line {
        x(xTime)
        y(y1)

        color = Color.BLUE
        width = 2.0
    }
    line {
        x(xTime)
        y(y2)

        color = Color.RED
        width = 2.0
    }
}
```
